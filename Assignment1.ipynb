A bar chart showing the revenue generated by each dish for better visuali

```python
total_revenue_by_dish = data_orders.groupby('Dish Name')['Amount (USD)'].sum()

total_revenue_by_dish = total_revenue_by_dish.sort_values(ascending=False)
# Plotting the total revenue by dish in ascending order
plt.figure(figsize=(6,5))
plt.bar(total_revenue_by_dish.index, total_revenue_by_dish.values, color='lightblue')
plt.title('Total Revenue by Dish (Ascending Order)')
plt.xlabel('Dish Name')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

 **conclusion** : The dish generating the highest revenue is Spaghetti, followed by Grilled Chicken. This suggests that these dishes are the most popular or have a higher price point, contributing significantly to total revenue. A focus on promoting these dishes can maximize profits.

Visualize the number of orders for each time of day.

```python
time_of_day_orders = data_orders['Time of Day'].value_counts()

plt.figure(figsize=(4,5))
sns.barplot(x=time_of_day_orders.index, y=time_of_day_orders.values, palette='Blues_d')
plt.title('Orders by Time of Day', fontsize=16)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.show()
```

**Conclusion** Most orders occur during the Night, followed by Day. Morning sees the least activity. This highlights the importance of focusing operational resources and marketing efforts during nighttime.

**Meal type vs Average session rating**

```python
avg_rating_by_meal = data_sessions.groupby('Meal Type')['Session Rating'].mean()
plt.figure(figsize=(6,6))
plt.pie(avg_rating_by_meal.values, labels=avg_rating_by_meal.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'skyblue', 'lightgreen'])
plt.title('Meal Type Share of Average Session Ratings')
plt.show()
```

**Conclusion** : All meal types have similar average ratings (around 4), suggesting consistent customer satisfaction. However, improving minor aspects of each meal type could push these ratings higher.

**Canceled order vs Dish name**

```python
canceled_orders = data_orders[data_orders['Order Status'] == 'Canceled']

# Group by Dish Name and count canceled orders
canceled_count_by_dish = canceled_orders['Dish Name'].value_counts()
plt.figure(figsize=(5, 4))
plt.bar(canceled_count_by_dish.index, canceled_count_by_dish.values, color='pink', width=0.3)  # Adjusting the width
plt.title('Canceled Orders vs Dish Name')
plt.xlabel('Dish Name')
plt.ylabel('Number of Canceled Orders')
plt.xticks(rotation=45)
plt.show()
```

**Conclusion** : Grilled Chicken and Veggie Burger have the highest cancellations. Investigate issues such as preparation delays, pricing, or customer preferences that might be leading to these cancellations.

Dish name vs Average session rating that dish

```python
avg_rating_by_dish =data_sessions.groupby('Dish Name')['Session Rating'].mean()

plt.figure(figsize=(6,5))
sns.barplot(x=avg_rating_by_dish.index, y=avg_rating_by_dish.values, palette='coolwarm')
plt.title('Dish Name vs. Average Session Rating', fontsize=16)
plt.xlabel('Dish Name', fontsize=12)
plt.ylabel('Average Session Rating', fontsize=12)
plt.xticks(rotation=45)  # Rotate dish names for better readability
plt.show()
```

 **conclusion** : The plot will show the average session rating for each dish.
Higher ratings could indicate customer satisfaction with specific dishes (e.g., Spaghetti or Grilled Chicken might have a higher rating).
Dishes with lower ratings could help identify areas where improvements in meal quality or service are needed.

**order ststus vs dish name**

```python
order_status_counts = data_orders.groupby(['Dish Name', 'Order Status']).size().unstack()
# Plotting the order status for each dish name
order_status_counts.plot(kind='bar', stacked=True, figsize=(7,5), color=['pink', 'lightblue'])
plt.title('Order Status vs Dish Name')
plt.xlabel('Dish Name')
plt.ylabel('Count of Orders')
plt.xticks(rotation=45)
plt.show()
```

# **Revenue vs. Ratings**

```python
rated_orders = data_orders[data_orders['Rating'].notna()]
plt.figure(figsize=(6, 5))
plt.boxplot([rated_orders[rated_orders['Rating'] == rating]['Amount (USD)'] for rating in range(1, 6)],
            labels=range(1, 6), patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Revenue Distribution by Rating')
plt.xlabel('Rating')
plt.ylabel('Revenue (USD)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**Conclusion**:Shows the spread and outliers for revenue within each rating category, giving a sense of variability

**meal type vs total orders**

```python
total_orders_by_meal_type = data_orders['Meal Type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(data_orders['Meal Type'].value_counts(), labels=data_orders['Meal Type'].value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Meal Type vs Total Orders')
plt.show()
```

```python
location_orders = data_user.groupby('Location')['Total Orders'].sum()

# Plotting the bar chart for Total Orders by Location
plt.figure(figsize=(10, 6))
location_orders.sort_values().plot(kind='bar', color='skyblue')
plt.title('Total Orders by Location')
plt.xlabel('Location')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```
